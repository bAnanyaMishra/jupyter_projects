# What are the locations with the fastest rollercoaster(min of 10)?

```python
#df2.query("Location!='Other'")
df2["Location"].value_counts()
```

```python
ax=df2.query("Location!='Other'").groupby('Location')['Speed_mph']\
 .agg(['mean','count']).query('count>=10')\
 .sort_values('mean')['mean']\
.plot(kind='barh',figsize=(10,5),title='Top 10 Average rollercoaster speed by location')
ax.set_xlabel('Average rollercoaster speed')
plt.show()
```
