<h2> Insights on data science dataset</h2>


```python
def details(df):
    print(red('Information on data science data set','bold'))
    print(blue('Total number of jobs in datascience '),end="")
    print(df.shape)
    print(blue("Information on the columns"))
    print(df.info())
    print(blue("Information on any null values present in the data"))
    print(df.isnull().sum())
```

```python
details(df)
```

```python
def duplicate_df(df):
    if(df.duplicated().sum()>0):
        print(red("There are few duplicates in the dataset",df.duplicated().sum))
        df.drop_duplicates(inplace = true)
        print(blue("details of the dataset after dropping duplicates"))
        details(df)
    else:
        print(green('There are no duplicates in the dataset'))
```

```python
duplicate_df(df)
```

```python
df.columns
```

```python
df['Job Title'].unique().tolist()
```

```python
#checking each column for unique values 

def unique_df(df):
    print("unique values in each column are")
    for i in df.columns:
        print(blue("total number of unique values in each column"),df[i].nunique())
        print(green("unique values in each column"),'\n',df[i].value_counts())  #value_counts sorts in descending order by default
```

```python
unique_df(df)
```

```python
df.columns
```

```python
#more insights 
def insights_df(df):
    print(blue('Total number of companies'),df['Company'].nunique())
    print(blue('Total number of Industries'),df['Industry'].nunique())
    print(blue('Total number of full time jobs'),df['Job Type'].isin(['Full Time']).sum())
    print(blue('Total number of Interships'),df['Job Type'].str.contains('Internship').sum())
    print(blue('Total number of Part time jobs'),df['Job Type'].str.contains('Part Time').sum())
    print(blue('Total number of Entry level jobs'),df['Career Level'].isin(['Entry']).sum())
    print(blue('Total number of Senior level jobs'),df['Career Level'].isin(['Senior']).sum())
    print(blue('Total number of Middle level jobs'),df['Career Level'].isin(['Middle']).sum())
    print(blue('Total number of degree qualifications'),df['Qualification'].isin(['Degree']).sum())
    print(blue('Total number of Postgraduate qualifications'),df['Qualification'].isin(['Postgraduate']).sum())
    print(blue('Total number of Buisness Analyst jobs'),df['Job Title'].str.upper().str.contains('BUSINESS ANALYST').sum())
    print(blue('Total number of Data Analyst jobs'),df['Job Title'].str.upper().str.contains('DATA ANALYST').sum())
    print(blue('Total number of Data Scientist jobs'),df['Job Title'].str.upper().str.contains('DATA SCIENTIST').sum())
    print(blue('Total number of ML Scientist jobs'),df['Job Title'].str.upper().str.contains('MACHINE LEARNING').sum())
```
